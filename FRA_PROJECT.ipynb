```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming you have prediction results for each model on the test set
test_predictions = {
    'Logistic Regression': pred_test,
    'Logistic Regression-SMOTE': pred_test_smote,
    'Random Forest': pred_test_rf,
    'Random Forest - Grid Search':  best_model_predictions,
    'LDA': lda_predictions,
    'LDA-SMOTE': lda_predictions_resampled,
}

# Calculate and fill in the precision, recall, and F1-score for each model on the test set
results = []

for model_name, test_preds in test_predictions.items():
    
    test_precision = precision_score(y_test, test_preds)
    test_recall = recall_score(y_test, test_preds)
    test_f1 = f1_score(y_test, test_preds)
    
    results.append([model_name, test_precision, test_recall, test_f1])

# Print the table for test set only
print("| Model                   | Test Precision | Test Recall | Test F1 Score |")
print("|-------------------------|----------------|-------------|--------------|")
for row in results:
    print("| {:<24} | {:<14.2f} | {:<11.2f} | {:<12.2f} |".format(row[0], row[1], row[2], row[3]))
```
